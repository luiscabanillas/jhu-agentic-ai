```python
from pathlib import Path
import os

def create_and_enter_directory(dir_name: str) -> Path:
    try:
        target_dir = Path(dir_name).resolve()

        # Create the directory if missing. No error if it already exists.
        target_dir.mkdir(parents=True, exist_ok=True)

        # Safety check: path must exist and must be a directory.
        if not target_dir.exists():
            raise FileNotFoundError(f"Directory was not created: {target_dir}")

        if not target_dir.is_dir():
            raise NotADirectoryError(f"Path exists but is not a directory: {target_dir}")

        # Change working directory.
        os.chdir(target_dir)

        # Safety check: confirm cwd changed as expected.
        current_dir = Path.cwd().resolve()
        if current_dir != target_dir:
            raise RuntimeError(
                f"Working directory mismatch. Expected: {target_dir}, Actual: {current_dir}"
            )

        print(f"Ready: using directory {current_dir}")
        return current_dir

    except PermissionError as err:
        print(f"Permission denied while creating or entering directory '{dir_name}': {err}")
        raise
    except OSError as err:
        print(f"OS error while creating or entering directory '{dir_name}': {err}")
        raise
    except Exception as err:
        print(f"Unexpected error for directory '{dir_name}': {err}")
        raise


create_and_enter_directory("PythonClass")
```

```python
from pathlib import Path
import os

def go_back_to_script_root() -> Path:
    os.chdir(START_DIR)

    current_dir = Path.cwd().resolve()
    if current_dir != START_DIR:
        raise RuntimeError(
            f"Failed to return to script root. Expected: {START_DIR}, Actual: {current_dir}"
        )

    print(f"Returned to script root: {current_dir}")
    return current_dir

go_back_to_script_root()
```
